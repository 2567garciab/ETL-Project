```python
import pandas as pd
```

```python
olympic_2008 = pd.read_csv('Resources/2008_data.csv', encoding='latin-1')
olympic_2008.head()
```

```python
new_olympic_2008 = olympic_2008[['Athlete', 'Event_gender', 'Country', 'Year', 'Sport', 'Event', 'Medal']]
new_olympic_2008.head()
```

```python
new2_olympic_2008 = new_olympic_2008.loc[new_olympic_2008['Year'] == 2008]
new2_olympic_2008.head()
```

```python
new2_olympic_2008["Year"] = pd.to_numeric(new2_olympic_2008['Year'], downcast='integer')
```

```python
new2_olympic_2008.head()
```

```python
new2_olympic_2008 = new2_olympic_2008.rename(columns={'Athlete':'Name', 'Event_gender':'Gender'})
new2_olympic_2008.head()
```

```python
new2_olympic_2008["First_Name"] = new2_olympic_2008["Name"].map(lambda x:x.strip().split(",")[-1])
```

```python
new2_olympic_2008["Last_Name"] = new2_olympic_2008["Name"].map(lambda x:x.strip().split(",")[0])
```

```python
new2_olympic_2008.head()
```

```python
new2_olympic_2008['Last_Name'] = new2_olympic_2008['Last_Name'].str.capitalize()
new2_olympic_2008.head()
```

```python
new2_olympic_2008['Event'] = new2_olympic_2008['Event'].str.title()
new2_olympic_2008.head()
```

```python
final_olympic_2008 = new2_olympic_2008[['First_Name', 'Last_Name', 'Gender', 'Country', 'Year', 'Sport', 'Event', 'Medal']]
final_olympic_2008 = final_olympic_2008.reset_index(drop=True)
final_olympic_2008.head()
```

```python
olympic_2012 = pd.read_csv('Resources/2012_data.csv')
olympic_2012.head()
```

```python
olympic_2016 = pd.read_csv('Resources/2012_data.csv')
olympic_2016.head()
```

```python
new_olympic_2012 = olympic_2012[['Name', 'Sex', 'Team', 'Year', 'Season', 'Sport', 'Event', 'Medal']]
new_olympic_2012.head()
```

```python
new_olympic_2016 = olympic_2016[['Name', 'Sex', 'Team', 'Year', 'Season', 'Sport', 'Event', 'Medal']]
new_olympic_2016.head()
```

```python
new2_olympic_2012 = new_olympic_2012.loc[(new_olympic_2012['Year']==2012) & (new_olympic_2012['Season'] == 'Summer')]
new2_olympic_2012.head()
```

```python
new2_olympic_2016 = new_olympic_2016.loc[(new_olympic_2016['Year']==2016) & (new_olympic_2016['Season'] == 'Summer')]
new2_olympic_2016.head()
```

```python
new2_olympic_2012 = new2_olympic_2012.dropna(how='any')
new2_olympic_2012.head()
```

```python
new2_olympic_2016 = new2_olympic_2016.dropna(how='any')
new2_olympic_2016.head()
```

```python
del new2_olympic_2012['Season']
new2_olympic_2012.head()
```

```python
del new2_olympic_2016['Season']
new2_olympic_2016.head()
```

```python
new2_olympic_2012 = new2_olympic_2012.rename(columns={'Sex':'Gender', 'Team':'Country'})
new2_olympic_2012.head()                                             
```

```python
new2_olympic_2016 = new2_olympic_2016.rename(columns={'Sex':'Gender', 'Team':'Country'})
new2_olympic_2016.head()   
```

```python
new2_olympic_2012["First_Name"] = new2_olympic_2012["Name"].map(lambda x:x.strip().split(" ")[0])
```

```python
new2_olympic_2012["Last_Name"] = new2_olympic_2012["Name"].map(lambda x:x.strip().split(" ")[-1])
```

```python
new2_olympic_2012.head()
```

```python
new2_olympic_2016["First_Name"] = new2_olympic_2016["Name"].map(lambda x:x.strip().split(" ")[0])
new2_olympic_2016["Last_Name"] = new2_olympic_2016["Name"].map(lambda x:x.strip().split(" ")[-1])
new2_olympic_2016.head()
```

```python
final_olympic_2012 = new2_olympic_2012[['First_Name', 'Last_Name', 'Gender', 'Country', 'Year', 'Sport', 'Event', 'Medal']]
final_olympic_2012 = final_olympic_2012.reset_index(drop=True)
final_olympic_2012.head()
```

```python
final_olympic_2016 = new2_olympic_2016[['First_Name', 'Last_Name', 'Gender', 'Country', 'Year', 'Sport', 'Event', 'Medal']]
final_olympic_2016 = final_olympic_2016.reset_index(drop=True)
final_olympic_2016.head()
```
